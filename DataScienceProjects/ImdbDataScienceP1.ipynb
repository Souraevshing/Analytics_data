# Scatter Plots

```python
mat.figure(figsize=(14,7))
plot=sns.boxplot(data=data, x='Rating', y='Revenue_Millions',color='red')
mat.savefig("boxplot.png")
```

```python
mat.figure(figsize=(14,7))
plot=sns.regplot(data=data, x='Rating', y='Revenue_Millions')
sns.set(font_scale=1.0)
mat.plot()
mat.savefig('LinePlot.png')
```

```python
#Group-by
data.groupby(['Votes','Revenue_Millions']).size().unstack().plot(kind='bar', stacked=True, figsize=(10,8));
mat.xlabel('Votes');
mat.ylabel('Revenue_Millions');
mat.xticks(rotation=60);
mat.yticks(rotation=60);
sns.set(font_scale=0.8);
mat.show();
mat.savefig('stackedbarPlot.png');
```
